```python
import matplotlib.pyplot as plt

# Horários do dia (0h a 24h)
horarios = list(range(25))

# Temperaturas fictícias para cada hora do dia
temperaturas = [15, 16, 17, 18, 19, 21, 22, 23, 24, 26, 28, 30, 30, 29, 27, 25, 23, 22, 21, 20, 19, 18, 18, 18, 18]

# Usando um estilo moderno do seaborn para um gráfico mais bonito
plt.style.use('seaborn-v0_8-darkgrid')

# Criando o gráfico com uma cor vibrante (laranja)
plt.plot(horarios, temperaturas, marker='o', color='tomato', linestyle='-', linewidth=3, markersize=8)

# Adicionando título e rótulos aos eixos
plt.title('Evolução da Temperatura Durante o Dia', fontsize=16, fontweight='bold', color='darkred')
plt.xlabel('Hora do Dia (h)', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Temperatura (°C)', fontsize=14, fontweight='bold', color='darkblue')

# Ajustando a grade para ficar mais suave
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

# Melhorando o formato do eixo X (intervalos de 2h) para ficarem mais visíveis
plt.xticks(range(0, 25, 2), fontsize=12, color='darkgreen')

# Melhorando a visibilidade do eixo Y
plt.yticks(fontsize=12, color='darkgreen')

# Definindo o fundo do gráfico e dos eixos
plt.gca().set_facecolor('honeydew')

# Exibindo o gráfico com layout ajustado
plt.tight_layout()
plt.show()
```
